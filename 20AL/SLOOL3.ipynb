```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.polynomial.polynomial import Polynomial


lengths = [63.0, 63.0, 37.6, 63.0, 21.8, 30.0, 40.1, 45.1, 50.0, 58.1]
lengths_squared = [L**2 for L in lengths]  # Length squared in cm^2
mean_periods = [1.57, 1.553, 1.189, 1.750, 0.889, 1.065, 1.230, 1.325, 1.391, 1.514]

# Fit a linear model to the data (best fit)
p = Polynomial.fit(lengths_squared, mean_periods, 1)

# Generate best-fit line values
x_fit = np.linspace(min(lengths_squared), max(lengths_squared), 100)
y_fit = p(x_fit)

plt.figure(figsize=(8, 6))
plt.scatter(lengths_squared, mean_periods, color='blue', label='Data Points')

#Best fit line
plt.plot(x_fit, y_fit, color='orange', linestyle='-', label='Best Fit Line')

plt.xlabel('Length^2 of Pendulum (cm^2)')
plt.ylabel('Period of Pendulum (s)')
plt.title('Period (T) vs Length^2 (L^2) of Pendulum')
plt.grid(True)
plt.legend()

plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for lengths in centimeters
lengths_cm = [63.0, 63.0, 37.6, 63.0, 21.8, 30.0, 40.1, 45.1, 50.0, 58.1]
lengths_cm_squared = [L**2 for L in lengths_cm]  # Length squared in cm^2
mean_periods_updated = [1.57, 1.553, 1.189, 1.750, 0.889, 1.065, 1.230, 1.325, 1.391, 1.514]

# Theoretical periods calculated based on g = 9.81
g_theoretical = 9.81  
lengths_m = [L / 100 for L in lengths_cm]  
theoretical_periods = [2 * np.pi * np.sqrt(L / g_theoretical) for L in lengths_m]  

# Plot measured data points
plt.figure(figsize=(8, 6))
plt.scatter(lengths_cm_squared, mean_periods_updated, color='blue', label='Measured Data Points')

# Plot the best-fit line for measured data
x_fit = np.linspace(min(lengths_cm_squared), max(lengths_cm_squared), 100)
coeff = np.polyfit(lengths_cm_squared, mean_periods_updated, 1)
y_fit = np.polyval(coeff, x_fit)
plt.plot(x_fit, y_fit, color='green', linestyle='-', label='Best Fit Line (Measured)')

# Plot the single theoretical line
lengths_cm_squared_sorted = sorted(lengths_cm_squared)
theoretical_periods_sorted = [2 * np.pi * np.sqrt(L / 100 / g_theoretical) for L in sorted(lengths_cm)]  # Sort for proper plotting
plt.plot(lengths_cm_squared_sorted, theoretical_periods_sorted, color='orange', linestyle='--', label='Theoretical Period according to Length ^2')

# Add labels and title
plt.xlabel('Length^2 of Pendulum (cm^2)')
plt.ylabel('Period of Pendulum (s)')
plt.title('Period vs Length^2 (Measured vs Theoretical)')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()

```
