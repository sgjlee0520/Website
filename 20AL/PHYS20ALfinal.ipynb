```python
import numpy as np

# Given measurements of the diameter
diameters = np.array([
    3.640, 3.450, 3.472, 3.500, 3.520, 3.520, 3.530, 3.530,
    3.432, 3.540, 3.550, 3.550, 3.560, 3.560, 3.570, 3.572,
    3.582, 3.582
])

# Calculate the mean of the diameters
mean_diameter = np.mean(diameters)
mean_diameter_sig = f"{mean_diameter:.3g}"  

# Calculate the RMSD (Root Mean Square Deviation) 
rmsd_diameter = np.sqrt(np.sum((diameters - mean_diameter) ** 2) / (len(diameters) - 1))
rmsd_diameter_sig = f"{rmsd_diameter:.1g}"  

print(mean_diameter, rmsd_diameter)
print(mean_diameter_sig, rmsd_diameter_sig)
```
