```python
import matplotlib.pyplot as plt
import numpy as np

# Data for lengths in centimeters
lengths_cm = [23, 30, 37, 44, 52.5]
mean_periods = [0.929, 1.081, 1.204, 1.328, 1.434]  # Measured periods

# Square the periods to get T^2
mean_periods_squared = [T**2 for T in mean_periods]

# Convert lengths to meters
lengths_m = [L / 100 for L in lengths_cm]

# Ideal periods calculated using g = 9.81
g_theoretical = 9.81
ideal_periods = [2 * np.pi * np.sqrt(L / g_theoretical) for L in lengths_m]
ideal_periods_squared = [T**2 for T in ideal_periods]  # Ideal T^2

# Fit a best-fit line for the measured T^2 vs L data
coeff_T_squared = np.polyfit(lengths_m, mean_periods_squared, 1)
x_fit_T_squared = np.linspace(min(lengths_m), max(lengths_m), 100)
y_fit_T_squared = np.polyval(coeff_T_squared, x_fit_T_squared)
plt.figure(figsize=(8, 6))

# Plot the measured T^2 vs L
plt.scatter(lengths_m, mean_periods_squared, color='blue', label='Measured T^2')

# best-fit line for measured data
plt.plot(x_fit_T_squared, y_fit_T_squared, color='green', linestyle='-', label='Best Fit Line (Measured)')

#theoretical period line
plt.plot(lengths_m, ideal_periods_squared, color='orange', linestyle='--', label='Ideal Period Line')


plt.xlabel('Length of Pendulum (m)')
plt.ylabel('Period^2 of Pendulum (s^2)')
plt.title('Measured T^2 vs L with Ideal Period Line')
plt.grid(True)
plt.legend()
plt.show()
```
