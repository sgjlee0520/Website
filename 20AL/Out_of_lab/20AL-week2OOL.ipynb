```python
import numpy as np
import matplotlib.pyplot as plt

# Data for trials 1 through 14 with corresponding angles
trial_numbers_full = list(range(1, 15))  # Trials 1 through 14
angles_full = [20, 10, 15, 13, 14, 13.5, 13.5, 13.5, 14, 13.5, 14, 13, 13.5, 13.5]  # Corresponding angles

# Create the scatter plot for trials 1 through 14 with trial numbers labeled
plt.figure(figsize=(8, 6))
plt.scatter(trial_numbers_full, angles_full, color='blue', label='Angles for Trials 1-14')

# Add a dashed horizontal line for the average critical angle, set to 13.5
plt.axhline(y=13.5, color='red', linestyle='--', label='Average Critical Angle = 13.5Â°')

# Mark trial numbers on the plot
for i, txt in enumerate(trial_numbers_full):
    plt.annotate(txt, (trial_numbers_full[i], angles_full[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Set y-axis ticks with 0.5 unit intervals
plt.yticks(np.arange(min(angles_full) - 1, max(angles_full) + 1, 0.5))

# Labels and title
plt.xlabel('Trial Number')
plt.ylabel('Angle (degrees)')
plt.title('Scatter Plot of Angle vs. Trial # (Trials 1-14)')
plt.legend()

# Show the plot
plt.show()
```
