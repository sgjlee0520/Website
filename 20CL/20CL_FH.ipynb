### Exercise 8:

In my analysis, I chose to determine the excitation energy by measuring the voltage differences between consecutive minima, since this approach cancels out any potential offsets and more directly captures the energy lost in inelastic collisions. For example, from my fine-sweep data I obtained minima at approximately 7.75V, 12.7V, and 17.8V, so the voltage differences are roughly 4.95V and 5.1V. I then take the average of these differences, which is about 5.03V, and calculate the first excitation energy as $ E_{1}^{Hg} = e \times \Delta V_{\text{avg}} \approx (1.602 \times 10^{-19} \, \text{C}) \times (5.03 \, \text{V}) \approx 5.03 \, \text{eV} $.

I used $ e = 1.602 \times 10^{-19} \, \text{C} $ for this calculation since it is an SI unit.

### Exercise 10: 
Below is my analysis using the given data.

I used the predicted excitation energy of mercury as $E_{1,accepted} = 4.67\,eV$ and my measured value from the experiment as $E_{1,measured} \approx 5.03\,eV$.  
The absolute discrepancy between these values is  
$$\Delta E = |5.03\,eV - 4.67\,eV| = 0.36\,eV.$$  
To find the proportionate discrepancy, I calculated:  
$$\text{Fractional Discrepancy} = \frac{0.36\,eV}{4.67\,eV} \approx 0.077,$$  
which is approximately 7.7%.  
**Thus, the measured excitation energy is about 7.7% higher than the predicted value.**
