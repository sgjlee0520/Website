```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = np.array([2026, 2027, 2028, 2029, 2030])
revenue = np.array([0.22, 2.48, 17.06, 44.42, 87.19])
gross_margin_percentage = np.array([0.20, 0.62, 0.74, 0.83, 0.90]) # Convert percentages to decimals

# Calculate gross margin amount and remaining revenue
gross_margin_amount = revenue * gross_margin_percentage
remaining_revenue = revenue - gross_margin_amount

# Create a stacked bar chart
plt.figure(figsize=(10, 7))
bar_width = 0.5

# Plot the remaining revenue (bottom segment)
plt.bar(years, remaining_revenue, bar_width, label='Remaining Revenue', color='skyblue')

# Plot the gross margin amount (top segment), stacked on top of remaining revenue
plt.bar(years, gross_margin_amount, bar_width, bottom=remaining_revenue, label='Gross Margin', color='lightcoral')

# Annotate the total revenue on top of each bar
for i in range(len(years)):
    plt.text(years[i], revenue[i] * 1.02, f'{revenue[i]:.2f}M', ha='center', va='bottom', fontsize=12)

# Labels and title
plt.xlabel('Year', fontsize=14)
plt.ylabel('Revenue (Million USD)', fontsize=14)
plt.title('Revenue and Gross Margin Growth (20262030)', fontsize=16)

# Customize x-ticks
plt.xticks(years, [str(year) for year in years], fontweight='bold', fontsize=12)

# Optionally adjust y-tick font size
plt.yticks(fontsize=12)

# Grid and layout
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.legend(fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```
