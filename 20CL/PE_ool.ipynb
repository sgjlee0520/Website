### Exercise 11

I will determine the stopping voltage $V_s$ by identifying the voltage where the photocurrent first drops to 0 nA (or within the noise floor, e.g., ≤0.001 nA) for each filter combination, using fine scan data where available (e.g., 546.1 nm, ND=OPEN) and broad scan data otherwise, applying linear interpolation between points if the transition occurs between measured voltages, and ensuring consistency by accounting for noise and offset.

Linear interpolation is necessary to estimate $V_s$ more precisely when $I=0$ lies between two measured data points, improving accuracy over simply picking the nearest voltage. The direct method can be less accurate because it doesn’t account for the continuous nature of the photocurrent’s transition to zero, is more sensitive to noise, and may not capture the non-linear behavior near $V_s$.

### Equation for interpolation

$$V_s=V_1+(\frac{I_1}{I_1-I_2})\cdot(V_2-V_1)$$

For example, if $I = 0.003nA$ at $V=0.6845 V$ and $I = 0.000nA $ at $V=0.6915 V$, $V_1=0.6845V, I_1=0.003nA, V_2=0.6915 V, I_2 = 0.000nA$.

$$V_s=V_1+(\frac{I_1}{I_1-I_2})\cdot(V_2-V_1)=0.6845+(\frac{0.003}{0.003-0.000})\cdot(0.6915-0.6845)=0.6845+0.0070=0.6915V$$

### Exercise 14

Using data taken without a neurtral density filter, plot $V_s=\frac{h}{e}\nu-\phi$ and fit the data to the line.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Data from Exercises 12 and 13 (ND = OPEN)
frequencies = np.array([8.22e14, 6.88e14, 5.49e14, 5.20e14])  # nu in s^-1
stopping_voltages = np.array([1.58, 1.17, 0.692, 0.585])  # V_s in V

# Define the linear function to fit: V_s = (h/e) * nu - phi
def linear_func(nu, h_over_e, phi):
    return h_over_e * nu - phi

# Perform the fit
popt, pcov = curve_fit(linear_func, frequencies, stopping_voltages)
h_over_e, phi = popt
h_over_e_err, phi_err = np.sqrt(np.diag(pcov))  # Uncertainties from covariance matrix

# Generate points for the fitted line
nu_fit = np.linspace(min(frequencies) * 0.95, max(frequencies) * 1.05, 100)
V_s_fit = linear_func(nu_fit, h_over_e, phi)

# Create a well-formatted plot
plt.figure(figsize=(8, 6))
plt.scatter(frequencies, stopping_voltages, color='blue', label='Data (ND = OPEN)', zorder=5)
plt.plot(nu_fit, V_s_fit, color='red', label=f'Fit: $V_s = ({h_over_e:.2e}) \\nu - {phi:.3f}$', zorder=3)
plt.xlabel('Frequency, $\\nu$ (s$^{-1}$)')
plt.ylabel('Stopping Voltage, $V_s$ (V)')
plt.title('Stopping Voltage vs. Frequency (ND = OPEN)')
plt.grid(True, which='both', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()

# Save the plot to a file
plt.savefig('Vs_vs_nu_ND_OPEN.png')
print(f"Fit parameter h/e = {h_over_e:.2e} ± {h_over_e_err:.2e} V·s")
print(f"Fit parameter phi = {phi:.3f} ± {phi_err:.2f} V")
```

### Exercise 15

Plot $V_s$ (O.D.) for one of the interference filters (546.1 nm, as chosen in Exercise 10) and fit the data to a line. Determine the fit parameters and their uncertainties.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Data for 546.1 nm
optical_densities = np.array([0.0, 0.2, 0.3, 0.6, 0.9, 1.5])  # O.D. values
stopping_voltages = np.array([0.692, 0.689, 0.685, 0.670, 0.655, 0.606])  # V_s in V

# Define the linear function to fit: V_s = m * O.D. + b
def linear_func(od, m, b):
    return m * od + b

# Perform the fit
popt, pcov = curve_fit(linear_func, optical_densities, stopping_voltages)
m, b = popt
m_err, b_err = np.sqrt(np.diag(pcov))  # Uncertainties from covariance matrix

# Generate points for the fitted line
od_fit = np.linspace(min(optical_densities) - 0.1, max(optical_densities) + 0.1, 100)
vs_fit = linear_func(od_fit, m, b)

# Create the plot
plt.figure(figsize=(8, 6))
plt.scatter(optical_densities, stopping_voltages, color='blue', label='Data (546.1 nm)', zorder=5)
plt.plot(od_fit, vs_fit, color='red', label=f'Fit: $V_s = ({m:.3f}) \\cdot \\text{{O.D.}} + {b:.3f}$', zorder=3)
plt.xlabel('Optical Density (O.D.)')
plt.ylabel('Stopping Voltage, $V_s$ (V)')
plt.title('Stopping Voltage vs. Optical Density (546.1 nm)')
plt.grid(True, which='both', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()

# Save the plot
plt.savefig('Vs_vs_OD_546nm.png')

# Print the fit parameters and their uncertainties
print(f"Slope (m) = {m:.3f} ± {m_err:.3f}")
print(f"Intercept (b) = {b:.3f} ± {b_err:.3f}")
```

The slope is small and negative, indicating a slight decrease in $V_s$ with increasing O.D., which may be due to experimental factors such as stray light and photocathode sensitivity changes, but theoretically $V_s$ should be constant.

### Planck’s Constant $h$

From Exercise 14, the slope of $V_s(\nu)$ is $3.27\pm 0.25 \times 10^{-15}V\cdot s$. The elementary charge $e=1.602\times10^{-19}$ C.

$$h=(\frac{h}{e})\cdot e = (3.27\times10^{-15})\times(1.602\times10^{-19})=5.239\times10^{-34}J\cdot s$$

### Uncertainty in $\frac{h}{e}$

The uncertainty in $\frac{h}{e}$ is calculated using the error propagation formula for multiplication:

$$ \frac{\sigma_{h}}{h} = \sqrt{ \left( \frac{\sigma_{h/e}}{h/e} \right)^2 + \left( \frac{\sigma_e}{e} \right)^2 } $$

Assuming $\sigma_e$ is negligible (as (e) is a known constant), the uncertainty simplifies to:

$$ \sigma_h = \left( \frac{\sigma_{h/e}}{h/e} \right) \times h $$

Given $\sigma_{h/e} = 0.25 \times 10^{-15} , \text{V·s} $,  $h/e = 3.27 \times 10^{-15} , \text{V·s} $, and $ e = 1.602 \times 10^{-19}$ , $\text{C}$ :

$$ \sigma_h = (0.25 \times 10^{-15}) \times (1.602 \times 10^{-19}) = 0.400 \times 10^{-34} = 4.00 \times 10^{-35}  \text{J·s} $$

Thus, the value of Planck’s constant is:

$$ h = (5.24 \pm 0.40) \times 10^{-34}, \text{J·s} $$
